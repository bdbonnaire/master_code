## Exercice 1 

We have $f(x) = e^{i2\pi F_0 x}$ and a sampling step $a=$$1\over F_e$ where $F_e = 32$.  
Given the expression of $f$ the fourier transform should give us a peak at $F_0$.

### 1. Reduction of $S_N(\lambda)$

$S_N(\lambda)$ is a geometric sum from $-N$ to $N-1$. Doing the appropriate factorisation gives us

$$
\begin{align}
    S_N(\lambda) &= e^{-i2\pi Na (F_0 - \lambda)} \sum\limits_{n=0}^{2N-1} e^{i2\pi na (F_0 - \lambda)}\\
    &= e^{-i2\pi Na (F_0 - \lambda)} \frac{1 - e^{i2\pi 2Na (F_0 - \lambda)}}{1 - e^{i2\pi a (F_0 - \lambda)}}\\
    &=\frac{e^{i2\pi Na (F_0 - \lambda)} - e^{i2\pi Na (F_0 - \lambda)}}{e^{i\pi a (F_0 - \lambda)} - e^{i\pi a (F_0 - \lambda)}} \frac{2i}{2i} e^{i\pi (F_0 - \lambda)a}\\
    &= \frac{\sin(2\pi Na (F_0 - \lambda))}{\sin(\pi a (F_0 - \lambda))} e^{i\pi (F_0 - \lambda)a}
\end{align}
$$

And then introducing $a=$$1\over F_e$ and $f_0 =$$ F_0 \over F_e$ We get
$$
S_N(\lambda) = \frac{\sin(2\pi N (f_0 - \frac{\lambda}{F_e}))}{\sin(\pi (f_0 - \frac{\lambda}{F_e}))} e^{i\pi (f_0 - \frac{\lambda}{F_e})}
$$

```python
# Comparison between approximated and analytical solutions
Fe = 32
N = 16

SNlambda = lambda l, f0, Fe : np.sin(2*np.pi * N * (f0 - l/Fe)) * np.exp(-1j*np.pi*(f0-l/Fe)) / np.sin(np.pi * (f0 - l/Fe))

# Calculation of SN(k/T) using fft when F0=7 :

f = lambda x, F0 : np.exp(2j*np.pi*F0*x)

F0 = 7
temp_signal = f(np.arange(-N, N)/Fe, F0)

SNk = np.fft.fft(temp_signal)

## plots

X = np.linspace(0, Fe, 200)

plt.plot(np.arange(0, 2*N) / (2*N/Fe), np.abs(SNk), label=r"$S_N(\frac{k}{T})$")
plt.plot(X, np.abs(SNlambda(X, F0/Fe, Fe)), label=r"$S_N(\lambda)$")
plt.title(r"Comparison of $S_N(\lambda)$ and $S_N(\frac{k}{T})$ when $F_0 = 7$", fontsize=16)
plt.legend()
plt.tight_layout()
plt. show()

# Calculation of SN(k/T) using fft when f0=0.2 :
f0 = 0.2

F0 = f0 * Fe
temp_signal = f(np.arange(-N, N)/Fe, F0)

SNk = np.fft.fft(temp_signal)

## plots 

plt.plot(np.arange(0, 2*N) / (2*N/Fe), np.abs(SNk), label=r"$S_N(\frac{k}{T})$")
plt.plot(X, np.abs(SNlambda(X, f0, Fe)), label=r"$S_N(\lambda)$")
plt.title(r"Comparison of $S_N(\lambda)$ and $S_N(\frac{k}{T})$ when $f_0 = 0.2$", fontsize=16)
plt.legend()
plt.tight_layout()
plt. show()
```

* With $F_0$ the approximation $S_N(k/T)$ is flat, equal to 0 exept around 7 where a peak appears. The theoretical value $S_N(\lambda)$ oscillates before and after the peak.
* With $f_0$ the approximation better approximates the theoretical value by always getting the peaks of the oscillation. 

## Exercice 2

In this exercice we consider the discrete signals $f = \sum\limits_{n\in \mathbb{Z}} f_n \delta_{na}$ and $d = {1\over 4} (\delta_0 + \delta_1 + \delta_2 + \delta_3)$.  
We study the convolution $g = d \star f = \sum\limits_{n\in \mathbb{Z}} g_n \delta_{na}$.

### 1. Expression of $g_n$
The convolution product on distributions gives us 
$$
<d\star f, \phi> = {1\over 4}\sum\limits_{k=0}^3\sum\limits_{n\in \mathbb{Z}} f_n \phi((k+n)a)
$$
Which implies
$$
\begin{align}
    d\star f &= {1\over 4}\sum\limits_{k=0}^3\sum\limits_{n\in \mathbb{Z}} f_n \delta_{(k+n)a}\\
    &= {1\over 4} ( \sum\limits_{n\in \mathbb{Z}} f_n \delta_{na} + \sum\limits_{n\in \mathbb{Z}} f_n \delta_{(1+n)a} + \sum\limits_{n\in \mathbb{Z}} f_n \delta_{(2+n)a} + \sum\limits_{n\in \mathbb{Z}} f_n \delta_{(3+n)a}\\
    &= \sum\limits_{n \in \mathbb{Z}} \underbrace{{1\over 4} (f_n + f_{n-1} + f_{n-2} + f_{n-3})}_{g_n} \delta_n
\end{align}
$$
So $g_n = {1\over 4} (f_n + f_{n-1} + f_{n-2} + f_{n-3})$. By the N-periodicity of $f_n$, the N-periodicity of $g_n$ is immediate. 

### 2. Computation of $g$ using DFT and IDFT

The fourier transform of a convolution is the product of the fourier transforms of the operands i.e. $\widehat{d \star f} = \hat{d} \hat{f}$.  
Thus to obtain $g$ using DFT we calculate the DFT of $d$ and $f$ separately then apply the inverse transform on the product $\hat{d} \hat{f}$.

#### Coresponding program

```python
# Definition of the function to calculate the convolution between two function
def convolution(a, b):
    """
    Applies the convolution product a*b and returns it by first calculating the DFT of a and b,
    multiplying them together and returning the inverse DFT.
    
    Arguments :
        a : ndarray, first operand of the convolution
        b : ndarray, second operand of the convolution
    """
    a_hat = np.fft.fft(a)
    b_hat = np.fft.fft(b)
    c = np.fft.ifft(a_hat * b_hat)

    return c
```

### 3. Application

We consider $f(x) = \sin(2\pi x)$ on the interval $[0,2[$ sampled at a 20 Hz frequency.

The discrete signal obtained can thus be written $\sum\limits_{n=0}^{39} f(na)\delta_{na}$ where $a =$ $1\over 20$.  
$d$ remains the same as before.

```python
# Application of the previous program

# sampling rate
a = 1/20

# sampling of [0,2[
X = np.linspace(0,2,int(2/a))

# definition of f
f = lambda x : np.sin(2*np.pi*x)

# definition of d
# no matter the sampling, d will always be one fourth on the first 4 elements
d = np.zeros(int(2/a))
d[0:4]=1/4

g = convolution(d, f(X))

# Plotting shenanigans
fig = plt.figure(constrained_layout = True)
fig.set_size_inches(8,6)
spec = fig.add_gridspec(ncols=2, nrows=2, figure = fig)
f_ax1 = fig.add_subplot(spec[0,0])
f_ax1.set(title=r"Signal $f$", ylabel="Magnitude")

f_ax2 = fig.add_subplot(spec[0,1])
f_ax2.set(title=r"Signal $d$", ylabel="Magnitude")

f_ax3 = fig.add_subplot(spec[1,:])
f_ax3.set(title=r" Convolution $d \star f$", ylabel="Magnitude")

f_ax1.stem(X, f(X))
f_ax2.stem(X, d)
f_ax3.stem(X,g.real)

fig.suptitle("Convolution product $d\star f$", fontsize=16)
plt.show()
```

### 4. Deconvolution 
Using the same property we used in **2.2** and supposing $\hat{d} \ne 0$ everywhere, we have
$$
\hat{f} = {\hat{g} \over \hat{d}}
$$
And $f$ can be recovered using the inverse Fourier transform.

If $\hat{d}$ vanishes at some points we have a problem as $\hat{g} \over \hat{d}$ is undefined. One way to work with such a hurdle is to replace the zero value with something small but different from 0. This idea is expanded below with the introduction of the inverse filter.

#### deconvolution code

```python
def deconvolution(d, g, c) :
    """
    if g = f*d with f a signal, returns f using the dft method
    
    Parameters:
        d = ndarray : one of the operand of the convolution
        g = ndarray : the result of the convolution
        c = float   : cutoff point for the inverse filter
    Returns :
        f = ndarray : the second operand of the convolution
    """
    
    d_hat = np.fft.fft(d)
    g_hat = np.fft.fft(g)
    
    # applies the inverse filter to d_hat with precision c
    d_hat_inv = inverse(d_hat, c)
    
    f = np.fft.ifft(g_hat / d_hat_inv)
    
    return f
```

### 5. Inverse filter

#### Inverse filter implementation

```python
def inverse(f, c) :
    """
    Inverse filter meant to avoid undefined values when dividing.
    
    Parameters:
        f = ndarray : the function to inverse.
        c = float   : cutoff point. When |f| gets below this value, 1/c is returned.
    
    Returns :
        f_inv = ndarray : 1/f if |f| > c, 1/c otherwise.
    """
    
    N = len(f)
    f_inv = np.zeros(N)
    
    for i in range(N) :
        if abs(f[i]) > c :
            f_inv[i] = 1/f[i]
        else :
            f_inv[i] = 1/c
    
    return f_inv
```

#### Illustration

```python
# Illustration of the inverse filter using the absolute value function

X = np.linspace(-2, 2, 150)
abs_inv = inverse(abs(X), 1/1.5)
real_inv = inverse(abs(X), 0.1)

fig2, ax = plt.subplots()

ax.plot(X, abs(X), 'k-', label= r"$|x|$")
ax.plot(X, real_inv, 'C3--', label= r"$\frac{1}{|x|}$")
ax.plot(X, abs_inv, 'C9-', label= "inverse filter")
ax.set_ylim(-0.1, 2)
ax.grid(alpha=.35)
fig2.suptitle(r"Inverse filter applied to $\frac{1}{|x|}$", fontsize=16)
fig2.legend()
plt.show()
```
